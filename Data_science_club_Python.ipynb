## **Ch 5 - Classes and objects.**

Python is an object oriented programming language. This means that everything in Python is treated as an object. Now what the heck is an object?

In programming, an object is anything that consists of both DATA and FUNCTIONS. It models real life better.

For example, the object car could have data:
1. The licence plate
2. The owner's name
3. Fuel capacity, etc.
4. Make, model

It could also have functions:
1. Accelerate
2. Brake
